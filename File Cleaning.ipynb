```python
import os
import pandas as pd
from openpyxl import load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows

# Define directories and dictionary file
hemorrhage_dir = 'Hemorrhage Data'
extra_erne_dir = 'Extra ERNE'
dict_file = 'Boosted Data Dictionary.xlsx'

# Read the mapping dictionary
df_dict = pd.read_excel(dict_file)
hemorrhage_col = df_dict.columns[0]
extra_erne_col = df_dict.columns[1]

for idx, row in df_dict.iterrows():
    hemorrhage_file = os.path.join(hemorrhage_dir, str(row[hemorrhage_col]))
    extra_erne_file = os.path.join(extra_erne_dir, str(row[extra_erne_col]))

    # Check if both files exist
    if not (os.path.exists(hemorrhage_file) and os.path.exists(extra_erne_file)):
        print(f"Skipping: {hemorrhage_file} or {extra_erne_file} not found.")
        continue

    # Load the T0-1 sheet from hemorrhage file
    try:
        df_t01 = pd.read_excel(hemorrhage_file, sheet_name='T0-1')
    except Exception as e:
        print(f"Could not read 'T0-1' from {hemorrhage_file}: {e}")
        continue

    # Open the extra ERNE file and add/replace the T0-1 sheet
    try:
        with pd.ExcelWriter(extra_erne_file, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
            df_t01.to_excel(writer, sheet_name='T0-1', index=False)
        print(f"Copied 'T0-1' from {hemorrhage_file} to {extra_erne_file}")
    except Exception as e:
        print(f"Could not write 'T0-1' to {extra_erne_file}: {e}")
```
