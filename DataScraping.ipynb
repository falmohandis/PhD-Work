```python
import os
import pandas as pd
from openpyxl import load_workbook

# Define folders
hemorrhage_folder = os.path.join(os.getcwd(), 'Hemorrhage Data')
processed_folder = os.path.join(os.getcwd(), 'Extra ERNE')
output_folder = os.path.join(os.getcwd(), 'Merged_Output')
os.makedirs(output_folder, exist_ok=True)

# Load mapping
mapping_file = os.path.join(os.getcwd(), 'Boosted Data Dictionary.xlsx')
df_map = pd.read_excel(mapping_file)

for _, row in df_map.iterrows():
    hem_file = str(row['Hemorrhage Data']).strip() + '.xlsx'
    proc_file = str(row['Processed_ERNE_Data']).strip() + '.xlsx'

    hem_path = os.path.join(hemorrhage_folder, hem_file)
    proc_path = os.path.join(processed_folder, proc_file)

    if not os.path.exists(hem_path):
        print(f"Skipping: Hemorrhage file not found: {hem_file}")
        continue
    if not os.path.exists(proc_path):
        print(f"Skipping: Processed file not found: {proc_file}")
        continue

    # Load workbooks
    wb_hem = load_workbook(hem_path, data_only=True)
    wb_proc = load_workbook(proc_path)

    # Append T0-1 from Hemorrhage Data into Processed ERNE
    if 'T0-1' in wb_hem.sheetnames:
        ws_hem = wb_hem['T0-1']
        # Create new sheet in processed workbook
        ws_new = wb_proc.create_sheet('T0-1')
        for row in ws_hem.iter_rows(values_only=True):
            ws_new.append(row)
    else:
        print(f"T0-1 not found in {hem_file}")

    # Save to merged folder
    out_name = f"merged_{os.path.splitext(proc_file)[0]}.xlsx"
    out_path = os.path.join(output_folder, out_name)
    wb_proc.save(out_path)
    print(f"Saved: {out_path}")
```
