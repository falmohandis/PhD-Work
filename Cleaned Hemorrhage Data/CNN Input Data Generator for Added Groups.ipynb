```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('Experiment Tracker.xlsx', sheet_name=['T30 categorization', 'T60 categorization'])

# Categorization at time=30
group_10 = set(df['T30 categorization']['Group_10'].dropna())
group_20 = set(df['T30 categorization']['Group_20'].dropna())
group_30 = set(df['T30 categorization']['Group_30'].dropna())

# Categorization at time=60
full_occlusion = set(df['T60 categorization']['Full_Occlusion'].dropna())
partial_occlusion = set(df['T60 categorization']['Partial_Occlusion'].dropna())
no_occlusion = set(df['T60 categorization']['No_Occlusion'].dropna())

# Print or use the variables as needed
print("Categorization at time=30:")
print("Group 10:", group_10)
print("Group 20:", group_20)
print("Group 30:", group_30)

print("\nCategorization at time=60:")
print("Full Occlusion:", full_occlusion)
print("Partial Occlusion:", partial_occlusion)
print("No Occlusion:", no_occlusion)
```

```python

# Find overlapping subjects for each combination at time=0
# Combine the sets
t_0_data = group_10 | group_20 | group_30

# Convert the set to a sorted list
sorted_t_0_data = sorted(t_0_data)


# Print the combined set
print("\nAvailable data at time=0:", sorted_t_0_data)


print("\n10% Hemorrhage: ", group_10)
print("\n20% Hemorrhage: ", group_20)
print("\n30% Hemorrhage: ", group_30)

# Find overlapping subjects for each combination at time=60
hemorrhage_10_full_occlusion = group_10.intersection(full_occlusion)
hemorrhage_10_partial_occlusion = group_10.intersection(partial_occlusion)
hemorrhage_10_no_occlusion = group_10.intersection(no_occlusion)

hemorrhage_20_full_occlusion = group_20.intersection(full_occlusion)
hemorrhage_20_partial_occlusion = group_20.intersection(partial_occlusion)
hemorrhage_20_no_occlusion = group_20.intersection(no_occlusion)

hemorrhage_30_full_occlusion = group_30.intersection(full_occlusion)
hemorrhage_30_partial_occlusion = group_30.intersection(partial_occlusion)
hemorrhage_30_no_occlusion = group_30.intersection(no_occlusion)



# Display the results for time=60
print("\n10% Hemorrhage + Full Occlusion at time=60:", hemorrhage_10_full_occlusion)
print("10% Hemorrhage + Partial Occlusion at time=60:", hemorrhage_10_partial_occlusion)
print("10% Hemorrhage + No Occlusion at time=60:", hemorrhage_10_no_occlusion)

print("\n20% Hemorrhage + Full Occlusion at time=60:", hemorrhage_20_full_occlusion)
print("20% Hemorrhage + Partial Occlusion at time=60:", hemorrhage_20_partial_occlusion)
print("20% Hemorrhage + No Occlusion at time=60:", hemorrhage_20_no_occlusion)

print("\n30% Hemorrhage + Full Occlusion at time=60:", hemorrhage_30_full_occlusion)
print("30% Hemorrhage + Partial Occlusion at time=60:", hemorrhage_30_partial_occlusion)
print("30% Hemorrhage + No Occlusion at time=60:", hemorrhage_30_no_occlusion)
```

```python
# Define the paths and arrays
parent_folder_path = os.getcwd()  # Change this to the actual path
parent_folders = os.listdir(parent_folder_path)
```

```python
raw_data_list = []
max_len = 0
# Iterate through parent folders
for parent_folder in parent_folders:
    parent_folder_full_path = os.path.join(parent_folder_path, parent_folder)
    
    # Check if it's a directory
    if os.path.isdir(parent_folder_full_path):
        
        # Add a second layer of folders: in this case, I will just focus on time=30
        for second_layer_folder in ["T25-26"]:
            second_layer_full_path = os.path.join(parent_folder_full_path, second_layer_folder)
            
            # Check if it's a directory
            if os.path.isdir(second_layer_full_path):
                
                # Construct the file name using the naming convention
                file_name = f"Raw.xlsx"
                file_path = os.path.join(second_layer_full_path, file_name)

#                 print(file_path)
                # Check if the file exists
                if os.path.isfile(file_path):
                    # Read the Excel file
                    raw_flow_data = pd.read_excel(file_path)
#                     print(raw_flow_data)
                    max_len = max(max_len, max(raw_flow_data.apply(lambda row: len(row.dropna()), axis=1)))

max_len=max_len*2
print(max_len)
```

```python
def replicate_numbers(row):
    row_without_nan = row.dropna()  # Remove NaN values
#     repeated_row = np.tile(row_without_nan, len(row_without_nan))  # Replicate to ensure enough length
    repeated_row = np.tile(row_without_nan, max_len // len(row_without_nan) + 1)[:max_len]
    
    # Apply Savitzky-Golay filter
    smoothed_row = savgol_filter(repeated_row, window_length=25, polyorder=3)
    
    return pd.Series(smoothed_row)

def process_raw_data(sheet_name,parent_folders, group_10, group_20, group_30,
                     hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                     hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                     hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion):
    raw_data_list = []

    for parent_folder in parent_folders:
        parent_folder_full_path = os.path.join(parent_folder_path, parent_folder)

        if os.path.isdir(parent_folder_full_path):
            for second_layer_folder in ["T25-26"]:
                second_layer_full_path = os.path.join(parent_folder_full_path, second_layer_folder)

                if os.path.isdir(second_layer_full_path):
                    file_name = f"Raw.xlsx"
                    file_path = os.path.join(second_layer_full_path, file_name)

                    if os.path.isfile(file_path):
                        raw_flow_data = pd.read_excel(file_path,sheet_name)
#Replicating
                        raw_flow_data = raw_flow_data.apply(replicate_numbers, axis=1)
#Replicating            
                        raw_flow_data["Parent Folder"] = parent_folder

                        if parent_folder in group_10:
                            raw_flow_data["Hemorrhage Level"] = 10
#                             print('Added')
                        elif parent_folder in group_20:
                            raw_flow_data["Hemorrhage Level"] = 20
#                             print('Added')
                        elif parent_folder in group_30:
                            raw_flow_data["Hemorrhage Level"] = 30
#                             print('Added')
                        else:
                            print(parent_folder)
                            print("Not found in any bleed group")
                            continue

                        if parent_folder in hemorrhage_10_full_occlusion:
                            raw_flow_data["Treatment Group"] = '10% Full'
                        elif parent_folder in hemorrhage_10_partial_occlusion:
                            raw_flow_data["Treatment Group"] = '10% Partial'
                        elif parent_folder in hemorrhage_10_no_occlusion:
                            raw_flow_data["Treatment Group"] = '10% None'
                        elif parent_folder in hemorrhage_20_full_occlusion:
                            raw_flow_data["Treatment Group"] = '20% Full'
                        elif parent_folder in hemorrhage_20_partial_occlusion:
                            raw_flow_data["Treatment Group"] = '20% Partial'
                        elif parent_folder in hemorrhage_20_no_occlusion:
                            raw_flow_data["Treatment Group"] = '20% None'
                        elif parent_folder in hemorrhage_30_full_occlusion:
                            raw_flow_data["Treatment Group"] = '30% Full'
                        elif parent_folder in hemorrhage_30_partial_occlusion:
                            raw_flow_data["Treatment Group"] = '30% Partial'
                        elif parent_folder in hemorrhage_30_no_occlusion:
                            raw_flow_data["Treatment Group"] = '30% None'
                        else:
                            print(parent_folder)
                            print("Not found in any treatment group")
                            continue

                        raw_data_list.append(raw_flow_data)

            for second_layer_folder in ["T0-1"]:
                second_layer_full_path = os.path.join(parent_folder_full_path, second_layer_folder)

                if os.path.isdir(second_layer_full_path):
                    file_name = f"Raw.xlsx"
                    file_path = os.path.join(second_layer_full_path, file_name)

                    if os.path.isfile(file_path):
                        raw_flow_data = pd.read_excel(file_path,sheet_name)
#Replicating
                        raw_flow_data = raw_flow_data.apply(replicate_numbers, axis=1)
#Replicating            
                        raw_flow_data["Parent Folder"] = parent_folder

                        if parent_folder in group_10:
                            raw_flow_data["Hemorrhage Level"] = 0
#                             print('Added')
                        elif parent_folder in group_20:
                            raw_flow_data["Hemorrhage Level"] = 0
#                             print('Added')
                        elif parent_folder in group_30:
                            raw_flow_data["Hemorrhage Level"] = 0
#                             print('Added')
                        else:
                            print(parent_folder)
                            print("Not found in any bleed group")
                            continue

                        if parent_folder in hemorrhage_10_full_occlusion:
                            raw_flow_data["Treatment Group"] = '10% Full'
                        elif parent_folder in hemorrhage_10_partial_occlusion:
                            raw_flow_data["Treatment Group"] = '10% Partial'
                        elif parent_folder in hemorrhage_10_no_occlusion:
                            raw_flow_data["Treatment Group"] = '10% None'
                        elif parent_folder in hemorrhage_20_full_occlusion:
                            raw_flow_data["Treatment Group"] = '20% Full'
                        elif parent_folder in hemorrhage_20_partial_occlusion:
                            raw_flow_data["Treatment Group"] = '20% Partial'
                        elif parent_folder in hemorrhage_20_no_occlusion:
                            raw_flow_data["Treatment Group"] = '20% None'
                        elif parent_folder in hemorrhage_30_full_occlusion:
                            raw_flow_data["Treatment Group"] = '30% Full'
                        elif parent_folder in hemorrhage_30_partial_occlusion:
                            raw_flow_data["Treatment Group"] = '30% Partial'
                        elif parent_folder in hemorrhage_30_no_occlusion:
                            raw_flow_data["Treatment Group"] = '30% None'
                        else:
                            print(parent_folder)
                            print("Not found in any treatment group")
                            continue

                        raw_data_list.append(raw_flow_data)

#     print(raw_data_list)
#     print(raw_data_list[0])
    raw_data_array = pd.concat(raw_data_list, ignore_index=True)
    return raw_data_array
#     return raw_data_list
```

```python
data_sources=["Vol",
              "Press",
              "Asc flow",
              "Renal flow",
              "Brachiocephalic flow",
              "Dist flow",
              "Carotid flow",
              "Aoballoon flow",
              "Prox pressure",
              "Dist press",
              "Rtjug pressure"]

vol_raw_data = process_raw_data(data_sources[0],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

press_raw_data = process_raw_data(data_sources[1],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

ascf_raw_data = process_raw_data(data_sources[2],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

renf_raw_data = process_raw_data(data_sources[3],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

brachiof_raw_data = process_raw_data(data_sources[4],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

distf_raw_data = process_raw_data(data_sources[5],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

carotf_raw_data = process_raw_data(data_sources[6],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

aoballf_raw_data = process_raw_data(data_sources[7],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

proxp_raw_data = process_raw_data(data_sources[8],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

distp_raw_data = process_raw_data(data_sources[9],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)

rtjugp_raw_data = process_raw_data(data_sources[10],parent_folders, group_10, group_20, group_30,
                             hemorrhage_10_full_occlusion, hemorrhage_10_partial_occlusion, hemorrhage_10_no_occlusion,
                             hemorrhage_20_full_occlusion, hemorrhage_20_partial_occlusion, hemorrhage_20_no_occlusion,
                             hemorrhage_30_full_occlusion, hemorrhage_30_partial_occlusion, hemorrhage_30_no_occlusion)
```
