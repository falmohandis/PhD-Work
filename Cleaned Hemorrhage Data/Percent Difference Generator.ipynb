```python
import pandas as pd
import numpy as np

# Load the data
file_path = 'PV Performance.xlsx'
df = pd.read_excel(file_path)

# Drop specified columns (except the ones needed for identification)
id_columns = ['Parent Folder', 'Hemorrhage Level', 'Treatment Group', 'Gender', 'Weight (kg)', 'Time Point']

id_columns = ['Parent Folder', 'Hemorrhage Level', 'Treatment Group', 'Gender', 'Weight (kg)', 'Time Point', 'Boosted Hemorrhage', 'Original Folder', 'Original Hemorrhage', 'Sampled Time Point', 'Simulated Time Point', 'Hemorrhage Rate %/Time', 'Time Point Numeric']


# df = df.drop(columns=['Hemorrhage Level', 'Treatment Group', 'Gender', 'Weight (kg)'])

# Extract numeric part from 'Time Point' and sort
def extract_time_point(t):
    try:
        return int(t.split('T')[1].split('-')[0])
    except (IndexError, ValueError):
        return np.nan

df['Time Point Numeric'] = df['Time Point'].apply(extract_time_point)
df = df.sort_values(by='Time Point Numeric')

# Specify the Time Point to start with (e.g., T29-30)
start_time_point = 'T29-30'
start_time_numeric = extract_time_point(start_time_point)

# Create an empty list to hold the modified dataframes
df_percent_diff_list = []
df_fractional_increase_list = []

# Iterate over unique Parent Folders
for parent_folder in df['Parent Folder'].unique():
    # Subset the dataframe for the current parent folder
    df_parent = df[df['Parent Folder'] == parent_folder]
    
    # Isolate the data at the start time point
    df_start = df_parent[df_parent['Time Point'] == start_time_point]
    
    if not df_start.empty:
        # Prepare dataframes for percent difference and fractional increase
        df_percent_diff = df_parent.copy()
        df_fractional_increase = df_parent.copy()
        
        # Loop over all columns except the identification columns
        for col in df_parent.columns.difference(id_columns):
            # Convert the column to numeric, coercing errors to NaN
            df_percent_diff[col] = pd.to_numeric(df_percent_diff[col], errors='coerce')
            df_fractional_increase[col] = pd.to_numeric(df_fractional_increase[col], errors='coerce')
            
            start_value = pd.to_numeric(df_start[col].values[0], errors='coerce')  # Get the value at the start time point

            # Calculate percent differences
            if pd.notnull(start_value) and start_value != 0:
                df_percent_diff[col] = df_percent_diff[col].apply(
                    lambda x: 100 * (x - start_value) / start_value if pd.notnull(x) else None)
                
            # Calculate fractional increases
            if pd.notnull(start_value):
                df_fractional_increase[col] = df_fractional_increase[col].apply(
                    lambda x: x / start_value if pd.notnull(x) else None)

        # Append the modified dataframes to the lists
        df_percent_diff_list.append(df_percent_diff)
        df_fractional_increase_list.append(df_fractional_increase)

# Concatenate all the modified dataframes back together
df_percent_diff_final = pd.concat(df_percent_diff_list)
df_fractional_increase_final = pd.concat(df_fractional_increase_list)

# Save the final dataframes to Excel files
df_percent_diff_final.to_excel(f'PV_Percent_Difference_{start_time_point}.xlsx', index=False)
df_fractional_increase_final.to_excel(f'PV_Fractional_Increase_{start_time_point}.xlsx', index=False)
```
